```python
import pandas as pd
```

```python
path = r"C:\Users\h\OneDrive\Desktop\Machine-Learning-Algorithms\Supervised Learning\Logistic Regression\loan_approval\data\raw_data/loan_data.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df["Gender"] =df["Gender"].ffill()
```

```python
df["Dependents"] = df["Dependents"].replace('3',3)
```

```python
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])
```

```python
df["Self_Employed"] = df["Self_Employed"].map({"Yes":1,"No":0})
df["Self_Employed"] = df["Self_Employed"].fillna(0)
```

```python
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].mean())
```

```python
df["Loan_Amount_Term"] =df["Loan_Amount_Term"].bfill()
```

```python
df["Credit_History"] = df["Credit_History"].fillna(0.0)
```
